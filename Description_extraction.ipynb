```python
import requests
import re
import json
from contextlib import suppress

i=0
csv_scrape = "id\tJob_Details"
for url in data['job_link']:
    url = str(url)
    if url.startswith('https://sjobs'):
        r = requests.get(url)
        soup = BeautifulSoup(r.content, 'lxml')
        with suppress(ZeroDivisionError, TypeError):
            preload_json = soup.find(id='preLoadJSON')['value']
            y = json.loads(preload_json)
            if y["Jobdetails"] is not None:
                print(i)
                #csv_scrape = csv_scrape + " \n "
                #csv_scrape = csv_scrape + str(i) + "\t"
                data['description'][i] = y["Jobdetails"]["JobDetailQuestions"][8]['AnswerValue'] #Job Description
                #csv_scrape = csv_scrape + y["Jobdetails"]["JobDetailQuestions"][7]['AnswerValue'] + "<dbrk>" #Location
                #csv_scrape = csv_scrape + y["Jobdetails"]["JobDetailQuestions"][11]['AnswerValue'] + "<dbrk>" #Job type
                #csv_scrape = csv_scrape + y["Jobdetails"]["JobDetailQuestions"][12]['AnswerValue'] + "<dbrk>" #Job Area of Interest
                #csv_scrape = csv_scrape + y["Jobdetails"]["JobDetailQuestions"][14]['AnswerValue'] + "<dbrk>" #Job address
                #csv_scrape = csv_scrape + y["Jobdetails"]["JobDetailQuestions"][19]['AnswerValue'] + "<dbrk>" #Job work state
                #csv_scrape = csv_scrape + y["Jobdetails"]["JobDetailQuestions"][6]['AnswerValue'] #Job title
            i = i+1
```

## Cleaning the extracted dataset

```python
import pandas as pd
import numpy as np

data = pd.read_csv('Company_Dataset/jobs_details_extracted.csv')
data.head(3)
```

```python
data.columns
```

```python
data['Function'].unique()
```

```python
data.shape
```

```python
data.dropna()
data.shape
```

```python
data = data.dropna(subset=['description']) # dropping all rows without description
data.shape
```

```python
import re

data.description = data.description.apply(lambda x: re.sub('<[^<]+?>', '', x)) #cleaning html
data.head()
```

```python
data.description = data.description.apply(lambda x: re.sub('&nbsp', '', x)) #cleaning non breaking spaces
data.head()
```

```python
data['description']
```

**Text is ready for NLP.**

```python
data.to_csv('Company_Dataset/jobs_details_extracted.csv',index=False) 
```
