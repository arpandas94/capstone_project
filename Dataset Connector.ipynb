```python
import numpy as np
import pandas as pd

data_cluster = pd.read_excel("LDA_Scikit_Cluster_Labeled.xlsx")
data_others = pd.read_csv("Company_Dataset/dataset_others.csv")
data_others = data_others.dropna(subset=["company_Link"])
print(data_cluster.shape)
print(data_others.shape)
```

They have similar entries which means they could be concatenated without any issue.

```python
data_cluster.head()
```

```python
data_others['Clusters_Labelled'] = data_cluster['dominant_topic'].to_numpy() #to stop index alignment issue
data_others.head(3)
```

```python
data_others['Clusters_Labelled'].unique()
```

```python
job_cluster_dict = {
    0 : "Accounting/Auditing",
    1 : "Accounting/Auditing",
    2 : "Civil Construction",
    3 : "Advertising",
    4 : "Training/Internship",
    5 : "Training/Internship",
    6 : "Writing/Editing",
    7 : "Customer Service",
    8 : "Team manager",
    9 : "Customer Service"
}

data_others = data_others.replace({"Clusters_Labelled": job_cluster_dict})
data_others.head(3)
```

```python
data_others.to_csv("Company_Dataset/dataset_others.csv", index=False)
```

```python
#Final Dataset Creation
data_ml_labelled = pd.read_csv("Company_Dataset/dataset_ml_labeled.csv")
data_others['Clusters_Labelled'] = data_others['ML Labeled Function']
data_others = data_others.drop(['Clusters_Labelled'], axis=1)
frames = [data_ml_labelled, data_others]
data_final = pd.concat(frames)
data_final.shape
```
